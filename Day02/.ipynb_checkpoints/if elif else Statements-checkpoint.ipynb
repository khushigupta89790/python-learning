Elif Statement 

#Syntax

    #excecute some code

```python
hungry = False

if hungry:
    print('FEED ME!')
else:
    print('Im not hungry')
```

```python
loc = 'Bank' 

if loc == 'Auto Shop':
    print ('Cars are cool')
elif loc == 'Bank' :
    print( 'Money is cool')
else:
    print("I do not know much")
```

```python
name = "sammy"
if name == "frankie":
    print("Hello frankie!")
elif name == 'sammy':
    print("Hello sammy!")
else:
    print('What is your name?')
```
