```python
# Q18.

# Check whether a year is a leap year.
year = int(input('Enter the year'))
if (year%4 == 0 and year%100 != 0) or (year % 400 == 0):
    print(f'{year} is leap year')
else:
    print(f'{year} is not leap year')
```

```python
# Q19.

# Input marks and print:

# 90+ → A
# 80–89 → B
# 70–79 → C
# 60–69 → D
# Below 60 → Fail
marks = int(input('Enter the marks'))
if marks >= 90:
    grade = 'A'
elif marks >= 80 :
    grade = 'B'
elif marks >= 70 :
    grade = 'C'
elif marks >= 60 :
    grade = 'D'
else :
    grade = 'Fail. Retry.'
print(f'Grade is {grade}')
```

```python
# Q24. Print numbers from 1 to 10 using a while loop.

z = 1
while z <= 10 :
    print(z)
    z += 1
```

```python
# Q25. Print numbers from 1 to 20, but stop when the number becomes 11 using break.

z = 1
while z <= 20 :
    if z == 11 :
        break
    print(z)
    z += 1
```

```python
# Q26. Print numbers from 1 to 10, but skip 5 using continue.
z = 1
while z <= 10 :
    if z == 5:
        z += 1
        continue
    print(z)
    z += 1
```
