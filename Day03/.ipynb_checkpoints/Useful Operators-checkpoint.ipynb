enumerate

```python
index_count = 0

for letter in 'abcde' :
    print(f'At index {index_count} the letter is {letter}') 
    index_count += 1
```

```python
index_count = 0
word = 'abcde'
for letter in word :
    print(word[index_count])
    index_count += 1 
```

```python
word = 'abcde'

for item in enumerate(word):
    print(item)
```

```python
word = 'abcde'

for index,letter in enumerate(word):
    print(index)
    print(letter)
    print('\n')
```

zip

```python
mylist1 = [1,2,3]
mylist2 = ['a' , 'b', 'c']
mylist3 = [100,200,300]
for item in zip(mylist1 , mylist2, mylist3):
    print(item)
```

```python
mylist1 = [1,2,3,4,5,6]
mylist2 = ['a' , 'b', 'c']
mylist3 = [100,200,300]
for item in zip(mylist1 , mylist2, mylist3):
    print(item)
```

```python
list(zip(mylist1,mylist2))
```
