Interactions between Functions

```python
example = [1,2,3,4,5,6,7]
```

```python
from random import shuffle
```

```python
shuffle(example)
```

```python
example
```

```python
def shuffle_list(mylist):
    shuffle(mylist)
    return mylist
```

```python
result = shuffle_list(example)
```

```python
result
```

```python
mylist = ['','O','']
```

```python
shuffle_list(mylist)
```

```python
def player_guess():

    guess = ''

    while guess not in ['0','1','2']:
        guess = input("Pick a number: 0,1, or 2")
    return int(guess)
```

```python
player_guess()
```

```python
my_index = player_guess()
```

```python
my_index
```

```python
def check_guess(mylist,guess):
    if mylist[guess] == 'O':
        print('Correct!!')
    else:
        print("Wrong guess!")
        print(mylist)
```

```python
#INITIAL GUESS
mylist = ['','O','']

#SHUFFLE LIST
mixedup_list = shuffle_list(mylist)

# USER GUESS
guess = player_guess()

# CHECK GUESS
check_guess(mixedup_list,guess)
```
