Lambda Expressions, Map, and Filter Functions

Map funtion

```python
def square(num):
    return num**2
```

```python
my_nums = [1,2,3,4,5]
```

```python
for item in map(square,my_nums):
    print(item)
```

```python
list(map(square,my_nums))
```

```python
def splicer(mystring):
    if len(mystring)%2==0:
        return 'Even'
    else: 
        return mystring[0]
```

```python
names = ['Andy','Eve','Sally']
```

```python
list(map(splicer,names))
```

```python
def check_even(num):
    return num%2 ==0
```

```python
mynums = [1,2,3,4,5,6]
```

Filter function

```python
list(filter(check_even,mynums))
```

```python
Lambda Expressions
```

```python
lambda num : num**2
```

```python
square(3)
```

```python
list(map(lambda num:num**2,mynums))
```

```python
list(filter(lambda num: num%2 == 0, mynums))
```

```python
names
```

```python
list(map(lambda name:name[::-1],names))
```
