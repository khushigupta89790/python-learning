```python
# Q1. Print all the words that end with 'e'.
# st = "Take the cake before everyone leaves the table"
st = "Take the cake before everyone leaves the table"
for word in st.split():
    if word[-1]=='e':
        print(word)
```

```python
# Q2. Print all numbers from 15 to 30 using range().
for x in range(15,31):
    print(x)
```

```python
# Q3. Print all odd numbers from 1 to 25.
for x in range(1,26):
    if x%2!=0:
        print(x)
```

```python
# Q4. Print the square of every number from 1 to 10.
for x in range(1,11):
    print(x**2)
```

```python
# Q5. Print only numbers divisible by 4 between 1 and 40.
for x in range(4,41,4):
    print(x)
```

```python
# Q6.
# st = "Python programming is super simple and so satisfying"

# Print only the words that start with 'p' or 'P'.
st = "Python programming is super simple and so satisfying"
for word in st.split():
    if word[0].lower()=='p':
        print(word)
```

```python
# Q7.

# Print each character of this string on a new line.

# word = "Developer"
word = "Developer"
for letter in word:
    print(letter)
    print('\n')
```

```python
# Q8.

# Using range(), print:

# 20
# 18
# 16
# 14
# 12
# 10
# 8
# 6
# 4
# 2
# 0
for num in range(20,-1,-2):
    print(num)
```

```python
# Q9.

# Create a list of all numbers between 1 and 30 divisible by 5 using list comprehension.
list(x for x in range(1,31) if x%5==0)
```

```python
# Q10.

# Create a list containing the cube of numbers from 1 to 10.
list(x**3 for x in range(1,11))
```

```python
# Q11.
# st = "sun moon star sky sea sand stone"

# Print only the words whose length is greater than 3.
st = "sun moon star sky sea sand stone"
for word in st.split():
    if len(word)>3:
        print(word)
```

```python
# Q12.

# Create a list of all even numbers from 1 to 50, but don't use if
list(x for x in range(2,51,2))
```

```python
# Q13.
# nums = [3, 8, 15, 20, 27, 32]

# Print only the even numbers.
nums = [3, 8, 15, 20, 27, 32]
print(list(x for x in nums if x%2 == 0))
```

```python
# Q14.

# Create a list of numbers from 1 to 100 divisible by 3 and 5.
[x for x in range (1,101) if x%3==0 and x%5==0]
```

```python
# Q15.
# sentence = "She sells sea shells on the sea shore"

# Print only the words that:

# start with 's' or 'S'
# and have more than 3 letters
sentence = "She sells sea shells on the sea shore"
for word in sentence.split():
    if word[0].lower()=='s' and len(word)>3 :
        print(word)
```
