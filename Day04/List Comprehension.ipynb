List Comprehension

```python
mystring = 'hello'
```

```python
mylist = []

for letter in mystring:
    mylist.append(letter)
```

```python
mylist
```

```python
mylist = [letter for letter in mystring]
```

```python
mylist
```

```python
mylist = [qweqwe for qweqwe in 'wordtwo']
```

```python
mylist
```

```python
mylist = [x**2 for x in range(0,11)]
```

```python
mylist
```

```python
mylist = [x**2 for x in range(0,11) if x%2 == 0]
```

```python
mylist
```

```python
celcius = [0,10,20,34.5]
farheneit = [( (9/5)*temp + 32) for temp in celcius]
```

```python
farheneit
```

```python
results = [x if x%2==0 else 'ODD' for x in range(0,11)] #less readable
```

```python
results
```

```python
mylist = []

for x in [2,4,6]:
    for y in [1,10,1000]:
        mylist.append(x*y)
```

```python
mylist
```

```python
mylist = [x*y for x in [2,4,6] for y in [1,10,1000]] #less readable
```

```python
mylist
```
